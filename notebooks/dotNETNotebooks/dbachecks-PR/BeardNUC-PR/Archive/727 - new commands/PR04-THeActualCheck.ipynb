# Convert-DbcResult

This command will take the results of `Invoke-DbcCheck` **IF** you use the `-PassThru` parameter and convert it into a data table for further processing - It will also do some of the data manipulation to make life easier for the (New) PowerBi

It will return these headings

Date Label Describe Context Name Database ComputerName Instance Result FailureMessage

You can specify the label to identify different types of tests or enable filtering in the final data set. You could use Prod, Test, Dev or Morning, Weekly, Monthly or System1, System2, System3. Which ever way that you want

```powershell
Invoke-DbcCheck -SqlInstance SQL2017N5 -Check LastGoodCheckDb -Show Summary -PassThru | Convert-DbcResult -Label BeardTest
```

Lets run with multiple instances

```powershell
Invoke-DbcCheck -SqlInstance SQL2017N5,SQL2019N5 -Check LastGoodCheckDb -Show Summary -PassThru | Convert-DbcResult
```

# Set-DbcFile

This command will output the dbachecks results as a CSV, XML or JSON file once they have been converted with `Convert-DbcResult`

Note - Update-DbcPowerBiDataSource outputs a json file but it does no data manipulation 

```powershell
$InvokeDbcCheck = @{
  SqlInstance = 'SQL2017N5','SQL2019N5' 
  Check = 'LastGoodCheckDb' 
  Show = 'Summary' 
  PassThru = $true
}
$SetDbcFile = @{
    FilePath =  'c:\temp'
    FileName ='SomeChecks.csv'
    FileType = 'Csv'
} 
Invoke-DbcCheck @InvokeDbcCheck | 
Convert-DbcResult | 
Set-DbcFile @SetDbcFile
```

# Write-DbcTable

We can also put the datatable into a database quickly with `Write-DbcTable`

I have no idea why this fails in a notebook as it works perfectly outside one - even on pwsh 7

```powershell
$InvokeDbcCheck = @{
    SqlInstance = 'SQL2017N5','SQL2019N5' 
    Check = 'LastGoodCheckDb' 
    Show = 'Summary' 
    PassThru = $true
  }
  $WriteDbcTable = @{
    SqlInstance = 'SQL2017N5'
    Database = 'tempdb'
    Table =  'dbacheck'
  } 
  Invoke-DbcCheck @InvokeDbcCheck | 
  Convert-DbcResult | 
  Write-DbcTable @WriteDbcTable -Verbose
```
